# Check whether female and male students scored the same marks.


H0:Null Hypothesis- female and male students scored the same marks

H1:Alternate Hypothesis-female and male students scored different marks

```python
female_data=data[data['Gender']=='female']
female_data
```

```python
female=female_data.drop(['Test preparation','Gender'],axis=1)
female
```

```python
male_data=data[data['Gender']=='male']
```

```python
male_data
```

```python
male=male_data.drop(['Test preparation','Gender'],axis=1)
male
```

```python
import scipy
tvalue, pvalue = scipy.stats.ttest_ind(male,female)
print('tvalue=',tvalue)
print('pvalue=',pvalue)
```

To find tcritical
scipy.stats.t.ppf(q, df)
where:

q: The significance level to use
df: The degrees of freedom

Here two tailed t test is used


```python
tcritical=scipy.stats.t.ppf(q=1-.05/2,df=26)
tcritical
```

If the absolute value of the test statistic is greater than the T critical value, then the results of the test are statistically significant(unlikely to have occurred given the null hypothesis)

```python
if np.abs(tvalue)>np.abs(tcritical) and pvalue <0.05:
    print('Reject null hypothesis that is Female and Male score equal marks')
else:
    print('Reject Alternate Hypothesis that is Female and Male does not score equal marks')
```

# Whether test preparation helps the students?

Null Hypothesis(H0):Test preparation helps the students
Alternate hypothesis(H1):Test preparation doesn't help the students

```python
test_prepared=data[data['Test preparation']=='completed']
test_prepared
```

```python
test_prepared1=test_prepared.drop(['Test preparation','Gender'],axis=1)
test_prepared1
```

```python
test_notprepared=data[data['Test preparation']=='none']
test_notprepared
```

```python
test_notprepared1=test_notprepared.drop(['Test preparation','Gender'],axis=1)
test_notprepared1
```

```python
tvalue, pvalue = scipy.stats.ttest_ind(test_prepared1,test_notprepared1)
print('tvalue=',tvalue)
print('pvalue=',pvalue)
```

```python
tcritical=scipy.stats.t.ppf(q=1-.05,df=26)
tcritical
```

```python
if np.abs(tvalue)>np.abs(tcritical) and pvalue <0.05:
    print('Reject Null hypothesis(H0):Test preparation helps the students')
else:
    print('Reject Alternate Hypothesis(H1):Test preparation doesnt helps the students ')
```

Test preparation didn't help the students
